# 전처리 함수 생성 및 전처리 진행

```python
def cleasing(text):
    repl =''
    pattern = '([ㄱ-ㅎㅏ-ㅣ]+)' # 자음, 모음 제거
    text = re.sub(pattern= pattern, repl=repl, string=text)
    #pattern = '[^\w\s]' # 특수기호 제거
    pattern = '[^가-히\s]' # 특수기호 제거
    text = re.sub(pattern= pattern, repl=repl, string=text)
    pattern = '<[^>]*>' # html 제거
    text = re.sub(pattern = pattern, repl='',string=text)
    
    pattern = ' +'
    text = re.sub(pattern = pattern, repl=' ', string=text)
    
    pattern = '"'
    text = re.sub(pattern = pattern, repl='', string=text)
    
    pattern = "'"
    text = re.sub(pattern=pattern, repl='', string=text)
    
    pattern = ","
    text = re.sub(pattern=pattern, repl='', string=text)
    
    text = text.strip()
    
    return text
```

```python
# 'document' 열에서 NaN 값을 빈 문자열로 대체
df_medi['요약문_연구목표'] = df_medi['요약문_연구목표'].fillna('')

# 'document' 열의 값을 문자열로 변환
df_medi['요약문_연구목표'] = df_medi['요약문_연구목표'].astype(str)

# 'document' 열에서 NaN 값을 빈 문자열로 대체
df_inf['요약문_연구목표'] = df_inf['요약문_연구목표'].fillna('')

# 'document' 열의 값을 문자열로 변환
df_inf['요약문_연구목표'] = df_inf['요약문_연구목표'].astype(str)
```

# 평균벡터 구하고 코사인 유사도 정의

```python
# 이전 데이터의 평균벡터
inf_vector_avg = np.mean(inf_df_embedding)
print(inf_vector_avg)
```

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm 

#코사인 유사도 계산
def cos_sim(vec1,vec2):
	return dot(vec1, vec2) / (norm(vec1)*norm(vec2))
```

```python
# 평균벡터와 이번에 학습할 데이터의 코사인 유사도 구하기
cos_rank = []
for i in range(len(medi_df_embedding)) :
    a = cos_sim(inf_vector_avg, medi_df_embedding[i])
    cos_rank.append(a)
```

```python
#코사인 유사도값 저장
# medi_df['cos_dis'] = cos_rank
medi_df['cos_dis'] = cos_rank
```
