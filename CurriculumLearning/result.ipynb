```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
full=pd.read_excel('/home/work/CL/기본추론결과.xlsx')
```

```python
plt.figure(figsize=(10,8))
plt.xlabel('Epoch')
plt.ylabel('ACC')

plt.plot(full['desc'],label='desc',color='limegreen',marker='o',linestyle='--')
plt.plot(full['asc'],label='asc', color='violet',marker='o',linestyle='--')
plt.plot(full['shuffle'],label='shuffle',color='dodgerblue',marker='o',linestyle='--')
# plt.plot(loss_dic['epoch'],nolora_df['train_loss'],label='no_lora_trainloss')
# plt.plot(loss_dic['epoch'],nolora_df['validation_loss'],label='no_lora_valloss')
plt.title("Full Fine-tuning ACC")
plt.legend()
plt.show()
```
